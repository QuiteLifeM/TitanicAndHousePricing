```python
import pandas as pd
import numpy as np
import config 
from sklearn.model_selection import KFold
```

```python
df = pd.read_csv(config.CONFIG['paths']['train_preprocessed'])
print(df.shape)
df.head()
```

```python
TARGET_COL = config.CONFIG['validation']['target_column']
N_SPLITS = config.CONFIG['validation']['n_splits']
RANDOM_STATE = config.CONFIG['validation']['random_state']
SHUFFLE = config.CONFIG['validation']['shuffle']

print('Таргет', TARGET_COL)
print('Фолдов', N_SPLITS)
```

```python
df = df.copy()
df['fold'] = -1
```

```python
kf = KFold(n_splits=N_SPLITS, shuffle=SHUFFLE, random_state=RANDOM_STATE)

for fold, (train_idx, val_idx) in enumerate(kf.split(df)):
    df.loc[val_idx, 'fold'] = fold
```

```python
df['fold'].value_counts().sort_index()
```

```python
output_path = config.CONFIG['paths']['train_with_folds']
df.to_csv(output_path, index=False)
print('Сохранено:', output_path)
print('Shape:', df.shape)
```
