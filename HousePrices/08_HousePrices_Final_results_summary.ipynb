# Итоговая сводка результатов

```python
import pandas as pd
import config

paths = config.CONFIG['paths']
all_dfs = []

if paths['metrics_results'].exists():
    df04 = pd.read_csv(paths['metrics_results'])
    df04['source'] = '04_classic_ML'
    all_dfs.append(df04)

if paths['dl_results'].exists():
    df05 = pd.read_csv(paths['dl_results'])
    df05['source'] = '05_DL'
    all_dfs.append(df05)

if paths['ensemble_results'].exists():
    df07 = pd.read_csv(paths['ensemble_results'])
    df07['source'] = '07_ensemble'
    all_dfs.append(df07)

if not all_dfs:
    print('Нет сохранённых результатов. Сначала запустите 04, 05 и 07 и сохраните метрики.')
else:
    summary = pd.concat(all_dfs, ignore_index=True)
    if 'std_rmse' not in summary.columns:
        summary['std_rmse'] = None
    summary = summary.sort_values('mean_rmse', ascending=True).reset_index(drop=True)
    summary['rank'] = range(1, len(summary) + 1)
    cols = ['rank', 'model', 'mean_rmse', 'std_rmse', 'source']
    summary = summary[[c for c in cols if c in summary.columns]]
    display(summary)
    summary.to_csv(paths['all_results'], index=False)
    print('Итог сохранён:', paths['all_results'])
```
